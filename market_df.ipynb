```python
#Histogram

plt.figure(figsize=(20,15))
for i,col in enumerate(num_df.columns,1):
    plt.subplot(5,5,i)
    sns.histplot(num_df[col], kde=True, bins=30,color='red')
    plt.title(col)
plt.tight_layout()
plt.show()

#The histogram plots with KDE curves indicate that several commodity and energy price variables are positively skewed with non-normal distributions, reflecting market volatility, price fluctuations,
#  and presence of extreme price spikes in certain sectors such as coal, gas, and oil markets.
```

```python
#Box plot

plt.figure(figsize=(27,7))
sns.boxplot(data=num_df,color='purple')
plt.title("Boxplot of Numerical columns")
plt.show()

#The boxplot reveals the presence of multiple outliers and significant variability in several commodity and market price columns,
# indicating high market volatility and extreme price fluctuations.
```

```python
#Line plot

plt.figure(figsize=(19,10))
for col in df.select_dtypes(include='number').columns:
    plt.plot(df.index, df[col], label=col)
plt.title("Line Plot of Numerical Columns")
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()

#The line plot shows significant volatility and fluctuating trends across commodity,energy,and market price variables over time, 
# with some variables exhibiting sharp spikes and extreme variations.
```

```python
#Heat map

plt.figure(figsize=(12,8))
sns.heatmap(num_df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()

#The correlation heatmap reveals strong positive correlations among coal, oil, LNG, and natural gas price variables, indicating that many commodity and energy markets tend to move together, while a few variables show weak or negative relationships,
# reflecting differing market behaviors and economic influences.
```
